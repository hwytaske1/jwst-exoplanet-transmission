The calibrated NIRISS/SOSS x1dints product contains extracted spectra for every integration during the observation. This section opens the FITS file and inspects its contents.

## 3. Construct the White-Light Pipeline

The `JWSTTransitPipeline` class loads the calibrated JWST spectral time series, stitches the observing segments together, removes bad pixels using the data quality (DQ) flags, and constructs a normalized white-light transit curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

class JWSTTransitPipeline:
    def __init__(self, filename):
        """Initializes the pipeline and loads the JWST FITS file."""
        self.filename = filename
        self.hdul = fits.open(filename)
        # SOSS Order 1 is sequentially stored in these extensions
        self.order1_exts = [2, 5, 8, 11]
        
        # Placeholders for processed data
        self.times = None
        self.flux_2d = None
        self.wave_1d = None
        self.dq_2d = None
        
    def load_and_stitch_segments(self):
        """Stitches the sequential file segments and filters out bad pixels."""
        segment_times = []
        segment_fluxes = []
        segment_waves = []
        segment_dq = []
        
        for ext_idx in self.order1_exts:
            ext_data = self.hdul[ext_idx].data
            segment_times.append(ext_data['TDB-MID'])
            segment_fluxes.append(ext_data['FLUX'])      
            segment_waves.append(ext_data['WAVELENGTH'])  
            segment_dq.append(ext_data['DQ'])
            
        # Concatenate across the time/integration axis
        self.times = np.concatenate(segment_times)
        self.flux_2d = np.concatenate(segment_fluxes, axis=0)
        self.dq_2d = np.concatenate(segment_dq, axis=0)
        
        # Take the median wavelength across time to get a static 1D wavelength grid
        self.wave_1d = np.nanmedian(np.concatenate(segment_waves, axis=0), axis=0)
        
        # Data Quality Cleaning: Mask out any pixel where DQ > 0 (bad pixels, cosmic rays)
        # We replace bad pixels with NaN so they are ignored during summation
        bad_pixel_mask = self.dq_2d > 0
        self.flux_2d[bad_pixel_mask] = np.nan
        
        print(f"Successfully stitched {len(self.times)} integrations across 2048 spectral channels.")
        print(f"Filtered out {np.sum(bad_pixel_mask)} bad pixel data points using DQ flags.")
        
    def get_white_light_curve(self):
        """Collapses the spectral axis to produce a normalized white-light curve."""
        if self.flux_2d is None:
            raise ValueError("Data not loaded. Run load_and_stitch_segments() first.")
            
        # Vectorized sum across the pixel axis (axis=1), safely ignoring NaNs
        white_flux = np.nansum(self.flux_2d, axis=1)
        
        # Normalize by median baseline
        oot = (self.times < 7.0) | (self.times > 7.12)
        normalized_flux = white_flux / np.nanmedian(white_flux[oot])
        return normalized_flux

    def get_spectral_light_curve(self, wave_min, wave_max):
        """
        Construct a light curve using only a selected wavelength range.

        Parameters
        ----------
        wave_min : float
            Minimum wavelength (microns)

        wave_max : float
            Maximum wavelength (microns)

        Returns
        -------
        normalized_flux : ndarray
            Normalized light curve
        """
    
        if self.flux_2d is None:
            raise ValueError("Data not loaded.")

        # Select wavelength channels
        wavelength_mask = (
            (self.wave_1d >= wave_min) &
            (self.wave_1d <= wave_max)
        )

        print(f"Using {np.sum(wavelength_mask)} wavelength pixels.")

        # Sum only those wavelengths
        spectral_flux = np.nansum(
            self.flux_2d[:, wavelength_mask],
            axis=1
        )

        # Normalize
        oot = (self.times < 7.0) | (self.times > 7.12)
    
        normalized_flux = spectral_flux / np.nanmedian(
            spectral_flux[oot]
        )

        return normalized_flux
    
    def close(self):
        """Closes the FITS file handler."""
        self.hdul.close()
```

## 4. Execute the Data Reduction Pipeline

This section loads the calibrated JWST observation, stitches the individual observing segments together, applies the data quality masks, and constructs a normalized white-light light curve.

```python
pipeline = JWSTTransitPipeline(
    "jw01366-o001_t001_niriss_clear-gr700xd-substrip256_x1dints.fits"
)

pipeline.load_and_stitch_segments()

norm_flux = pipeline.get_white_light_curve()
```

## 5. White-Light Transit Curve

The white-light curve is obtained by summing the stellar flux across all wavelength channels for each integration. The resulting time series reveals the planetary transit.

```python
plt.figure(figsize=(10,5))

plt.scatter(
    pipeline.times,
    norm_flux,
    s=10,
    color="black",
    alpha=0.6
)

plt.xlabel("Time (TDB - 59780 MJD)")
plt.ylabel("Normalized Flux")
plt.title("JWST WASP-39 b White-Light Transit")

plt.grid(alpha=0.3)

plt.show()
```

## 6. Transit Measurements

We estimate the average out-of-transit flux, the average in-transit flux, the transit depth, and the out-of-transit scatter.

```python
oot = (pipeline.times < 59787.0) | (pipeline.times > 59787.12)
transit = (pipeline.times >= 59787.0) & (pipeline.times <= 59787.12)

oot_mean = np.mean(norm_flux[oot])
transit_mean = np.mean(norm_flux[transit])

depth = oot_mean - transit_mean

scatter = np.std(norm_flux[oot])

print(f"Out-of-transit mean : {oot_mean:.6f}")
print(f"In-transit mean     : {transit_mean:.6f}")
print(f"Transit depth       : {depth:.6f}")
print(f"Scatter             : {scatter*1e6:.0f} ppm")
```
